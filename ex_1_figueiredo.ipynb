# Exercício 1 do Livro Estatística Descritiva e Probabilidade - Figueiredo


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_by_region = {
    "Alto Trás-os-Montes": 223333,
    "Ave": 509968,
    "Cávado": 393063,
    "Douro": 221853,
    "Entre Douro-e-Vouga": 276812,
    "Grande Porto": 1260680,
    "Minho-Lima": 250275,
    "Tâmega": 551309
}
```

```python
f,ax = plt.subplots(figsize=(18,5))
plt.bar(population_by_region.keys(), population_by_region.values()) # Gera o gráfico de barras
ax.set_ylim(100000, 1500000) # Limites em 100.000 e 1.500.000 para ter uma melhor visualização da diferença
plt.ticklabel_format(style='plain', axis='y') # Remove a notação científica - sem esta linha o eixo Y é representado em 1e6
plt.ylabel('Milhoes de habitantes')
plt.title('Distribuição de habitantes por região')
```

```python
labels = population_by_region.keys()
data = list(population_by_region.values())
highest_pop = max(data)
highest_pop_index = data.index(highest_pop)
explode_config = [0 if value != highest_pop else 0.15 for value in data]
fig1, ax1 = plt.subplots()

ax1.pie(data, labels=labels, autopct='%1.1f%%', explode=explode_config)
```
